issue 11- The none values in doggo, floofer, pupper, puppo columns were dropped

```python
dfArcClean[['doggo', 'floofer', 'pupper', 'puppo']].value_counts()
```

```python
#code
#list of dog names
stages = ['doggo', 'floofer', 'pupper', 'puppo']

#replacing none with nan
dfArcClean[stages] = dfArcClean[stages].replace('None', np.nan)

#function to join them
def join(x):
    return ', '.join(x.dropna().astype(str))
#applying the join
dfArcClean['stages'] = dfArcClean[stages].apply(join, axis=1)
#replacing empty string with nan
dfArcClean['stages'] = dfArcClean['stages'].replace('',np.nan)
```

```python
#test
dfArcClean.info()
```

issue The p1, p1_conf, p1_dog, p2, p2_conf, p2_dog, p3, p3_conf, and p3_dog columns were all combined into two columns breeds and confidence

```python
#code
condition = [(dfImgClean['p1_dog'] == True),(dfImgClean['p2_dog'] == True), (dfImgClean['p3_dog'] == True)]

dogBreed = [dfImgClean['p1'], dfImgClean['p2'], dfImgClean['p3']]

dogConf = [dfImgClean['p1_conf'], dfImgClean['p2_conf'], dfImgClean['p3_conf']]

dfImgClean['breed'] = np.select(condition, dogBreed, default = 'None')

dfImgClean['prediction'] = np.select(condition, dogConf, default = 0)
```

```python
#test
dfImgClean.info()
```
